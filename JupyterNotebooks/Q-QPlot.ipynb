##### Normal Distribution

```python
from statsmodels.graphics.gofplots import qqplot
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
color = sns.color_palette()
plt.rcParams["figure.figsize"]=50,50
%matplotlib inline
from scipy.stats import skewnorm
```

```python
import numpy
mu, sigma = 0, 1
samples = numpy.random.normal(mu, sigma, 10000)
```

```python
samples
```

```python
def Plot_Dist(data):
    fig,axarr = plt.subplots(1,2,figsize=(12,4))
    # plot distribution
    sns.distplot(data,ax=axarr[0])
    #Q-Q plot
    qqplot(data,line='s',ax=axarr[1])
    fig.suptitle('normal distribution',fontsize=14)
    plt.show()
```

```python
Plot_Dist(samples)
```

##### Non -Normal Distribution

```python

import numpy as np
from matplotlib import pyplot as plt
non_normal_data = skewnorm.rvs(10, size=1000)
```

```python
Plot_Dist(non_normal_data)
```
