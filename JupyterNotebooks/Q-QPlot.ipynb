```python
from statsmodels.graphics.gofplots import qqplot
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
color = sns.color_palette()
plt.rcParams["figure.figsize"]=50,50
%matplotlib inline
```

```python
import numpy
mu, sigma = 0, 1
samples = numpy.random.normal(mu, sigma, 10000)
```

```python
samples
```

```python
def Plot_Dist():
    fig,axarr = plt.subplots(1,2,figsize=(12,4))
    # plot distribution
    sns.distplot(samples, fit=norm, kde=False,ax=axarr[0])
    #Q-Q plot
    qqplot(samples,line='s',ax=axarr[1])
    fig.suptitle('normal distribution',fontsize=14)
    plt.show()
```

```python
Plot_Dist()
```
